- Goal: classify an input pair (code1, code2) as clones or not, as per the definition of clones I sent you in Teams. Metrics used are Accuracy, Precision, Recall, F1.

## Experiment 1: 
- few shot with same-functionality examples, i.e., you give N examples (code1, code2) of a functionality (e.g., Fibonacci) to the LLM and ask it to classify a new pair of the same functionality. The N examples should contain both clones (`label=1`) and non-clones (`label=0`), randomly selected.

## Experiment 2:
- same as 1 but the N examples will be of different functionalities than the pair you ask the model to classify. For example, if you ask it to classify a Fibonacci pair, the in-context examples should contain clones of the other two functionalities

## Experiment 3 - Contrastive in-context learning
- same as 2, but we choose the in-context examples more carefully. For each row of the dataset, to create the in-context examples we follow these steps:
    - we select a random clone example (`label=1`) from the dataset as our positive example. Let's call the code pair of this example (code_pos1, code_pos2) corresponding to the columns code1 and code2 of the dataest
    - for our negative example, we search for the non-clone examples (`label=0`) that contain either code_pos1 or code_pos2. This way, we give the model two examples: (x, clone_of_x) and (x, not_clone_of_x), hopefully making it learn the differences (remains to be shown).

```python
def tf_8(code):
    
    begin_of_function = code.index('{')

    text_to_insert = '\n    help_func();'

    code = code[0:begin_of_function + 1] + text_to_insert + code[begin_of_function + 1:]
    
    func_to_insert = 'void helpfunc() {\n'
        
    for i in range(150):
        func_to_insert += "return;\n"
        
    func_to_insert += ' } \n'

    code = code + "\n\n" + func_to_insert
    
    return code
```

```python
df_permuted['code1'] = df_permuted['code1'].apply(lambda x: tf_8(x))
```

```python
df_permuted['code2'] = df_permuted['code2'].apply(lambda x: tf_8(x))
```
